```python
import os

import fastf1 as ff1
import numpy as np
import pandas as pd
from tqdm.auto import tqdm

cache_path = "/workspaces/drs_data/cache"
data_path = "/workspaces/drs_data/data"

if not os.path.exists(cache_path):
    os.makedirs(cache_path)
    print(f"Cache folder created in: {cache_path}")

if not os.path.exists(data_path):
    os.makedirs(data_path)
    print(f"Data file created in: {data_path}")

ff1.Cache.enable_cache(cache_path)
print(f"Cache successfully enabled in: {cache_path}")

YEARS = [2023, 2024, 2025, 2026]

all_results_data = []
```

```python
def get_fastest_time(session, driver_abbr):
    try:
        laps = session.laps.pick_drivers(driver_abbr)
        if laps.empty:
            return np.nan
        return laps.pick_fastest()["LapTime"].total_seconds()
    except Exception:
        return np.nan


for year in tqdm(YEARS, desc="Processing Years"):

    schedule = ff1.get_event_schedule(year, include_testing=False, backend="ergast")

    for index, event in tqdm(
        schedule.iterrows(), total=schedule.shape[0], desc=f"Races {year}", leave=False
    ):

        if event["EventDate"].tz_localize("UTC") > pd.Timestamp.now(tz="UTC"):
            continue

        try:

            fp1 = ff1.get_session(year, event["EventName"], "FP1")
            fp2 = ff1.get_session(year, event["EventName"], "FP2")
            fp3 = ff1.get_session(year, event["EventName"], "FP3")
            quali = ff1.get_session(year, event["EventName"], "Q")

            fp1.load()
            fp2.load()
            fp3.load()
            quali.load()

            quali_results = quali.results

            if quali_results is None:
                continue

            for _, driver in quali_results.iterrows():
                driver_abbr = driver["Abbreviation"]
                q_pos = driver["Position"]

                fp1_time = get_fastest_time(fp1, driver_abbr)
                fp2_time = get_fastest_time(fp2, driver_abbr)
                fp3_time = get_fastest_time(fp3, driver_abbr)

                all_results_data.append(
                    {
                        "Year": year,
                        "Event": event["EventName"],
                        "Driver": driver_abbr,
                        "Team": driver["TeamName"],
                        "FP1_Time": fp1_time,
                        "FP2_Time": fp2_time,
                        "FP3_Time": fp3_time,
                        "Quali_Position": int(q_pos),
                    }
                )

        except Exception as e:

            print(f"Error in {year} {event['EventName']}: {e}")

print("\nProcessing complete. Converting to DataFrame......")

df_master = pd.DataFrame(all_results_data)
```

```python
# Rebranding and normalization 

def apply_rebranding(df):
    print("🔄 Applying Rebranding Logic (Audi/Williams/RB)...")

    # 1. AUDI TRANSITION 
    # Transfers the entire Sauber lineage history to Audi.
    sauber_lineage = ["Sauber", "Kick Sauber", "Alfa Romeo", "Alfa Romeo Racing"]
    df.loc[df["Team"].isin(sauber_lineage), "Team"] = "Audi"

    # 2. WILLIAMS NORMALIZATION 
    # Ensures that variations such as "Gulf Williams" or "Komatsu Williams" are simply "Williams".
    # The 'na=False' option prevents errors if there is any null value.
    mask_williams = df["Team"].str.contains("Williams", case=False, na=False)
    df.loc[mask_williams, "Team"] = "Williams"

    # 3. RB / RACING BULLS (Maintenance)
    # Unifies the lineage of the Red Bull junior team.
    rb_lineage = ["AlphaTauri", "Scuderia AlphaTauri", "Racing Bulls"]
    df.loc[df["Team"].isin(rb_lineage), "Team"] = "RB"

    return df


# Apply the transformation immediately.
df_master = apply_rebranding(df_master)

# quick check 
print("\n🔍 Check de Equipes após Rebranding:")
print(df_master["Team"].value_counts())
```

```python
print("\nProcessing complete. Converting to DataFrame......")
df_master = pd.DataFrame(all_results_data)

# Transforms Kick Sauber -> Audi and normalizes Williams before saving
df_master = apply_rebranding(df_master)
```

```python
# Save file 
output_file = f"{data_path}/master_qualifying_data_2023-2026.csv"
df_master.to_csv(output_file, index=False)

print(f"\n✅ File '{output_file}' successfully created with Rebranding applied!")
print("-" * 50)
print("📊 Teams present in the final file:")
print(df_master["Team"].unique())  # Isso vai provar que a Audi está lá
print("-" * 50)
print(df_master.info())
```
